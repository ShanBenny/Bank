```python
#Drop duplicate IFSC (keep first occurrence)
df = df.drop_duplicates(subset=["IFSC"], keep="first")
df["IFSC"].nunique()
```

```python
df = df.fillna("")
df
```

```python
# Clean STD CODE
df["STD CODE"] = (
    df["STD CODE"]
    .astype(str)
    .str.replace(".0", "", regex=False)
    .replace({"nan": "", "NaN": ""})
    .fillna("")
)

# Replace invalid STD values
df["STD CODE"] = df["STD CODE"].apply(lambda x: "NA" if x.strip() in ["", "0"] else x)

# Clean PHONE
df["PHONE"] = (
    df["PHONE"]
    .astype(str)
    .str.replace(".0", "", regex=False)
    .replace({"nan": "", "NaN": ""})
    .fillna("")
)

# Final check: replace blanks with NA
df["PHONE"] = df["PHONE"].apply(lambda x: "NA" if x.strip() == "" else x)
df[["STD CODE","PHONE"]]
```
