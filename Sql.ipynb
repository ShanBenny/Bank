```python
import psycopg2
import pandas as pd

# Load cleaned CSV
df = pd.read_csv("Cleaned.csv")

# Ensure column names match Postgres
df.columns = ["bank", "ifsc", "branch", "address", "city1", "city2", "state", "stdCode", "phone"]

# Connect to PostgreSQL
conn = psycopg2.connect(
    host="localhost",
    database="Bank",
    user="postgres",
    password="Shan",
    port="5432"
)
cur = conn.cursor()

# Insert rows (bank_id is auto-generated, so we skip it)
for _, row in df.iterrows():
    cur.execute("""
        INSERT INTO bank_details(bank, ifsc, branch, address, city1, city2, state, stdCode, phone)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
        ON CONFLICT (ifsc) DO NOTHING;
    """, (
        row["bank"],
        row["ifsc"],
        row["branch"],
        row["address"],
        row["city1"],
        row["city2"],
        row["state"],
        row["stdCode"],
        row["phone"]
    ))

conn.commit()
cur.close()
conn.close()

print("✅ Data inserted successfully,")
```
